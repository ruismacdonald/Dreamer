```python
for buffer, file in [("FIFO", "fifo"), ("LoFoV1", "v1"), ("LoFoV2", "v2")]:
    
    plt.rcParams.update({'font.size': 50})
    figure, axes = plt.subplots(1, 2, figsize=(30, 11))
    
    for ax, phase in zip(axes, [1, 2]):
        with open(f"/Users/ruismacdonald/Downloads/rewards/{file}_1_phase{phase}/heat_data.npy", "rb") as f:
            heat_data = np.load(f, allow_pickle=True)
            
        heat_data[heat_data[:,2] == 1, 2] = 4

        count = int(np.sqrt(heat_data.T[0].shape[0]))
        factor = 2.
        a = heat_data.T[0].reshape(count, count)
        b = heat_data.T[1].reshape(count, count)
        c = heat_data.T[2].reshape(count, count) / factor
        c = c[:-1, :-1]
        c[c==1] = 4.

        l_a, r_a = a.min(), a.max()
        l_b, r_b = b.min(), b.max()
        l_c, r_c = 0, 4

        mesh = ax.pcolormesh(a, b, c, cmap='hot', vmin=l_c, vmax=r_c)
        ax.set_title(f"{buffer} - Phase {phase}", size=55)
        ax.axis([l_a, r_a, l_b, r_b])
        figure.colorbar(mesh, ax=ax)

    plt.tight_layout()
    plt.savefig(f'/Users/ruismacdonald/Downloads/RHM_{file}_both_phases.pdf')
    plt.show()
```
