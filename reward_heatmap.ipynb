```python
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
import matplotlib

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = '15'
plt.rcParams['text.usetex'] = True
matplotlib.rcParams['axes.spines.right'] = False
matplotlib.rcParams['axes.spines.top'] = False


%matplotlib inline
```

```python
with open("/Users/ruismacdonald/Research/Dreamer/reward/reacherloca_optimal/heat_data_optimal.npy", "rb") as f:
    heat_data = np.load(f, allow_pickle=True)
    
heat_data[heat_data[:,2] == 1][:,2] = 4

title="{buffer} - Phase {phase}"
    
count = int(np.sqrt(heat_data.T[0].shape[0]))
factor = 2.
a = heat_data.T[0].reshape(count, count)
b = heat_data.T[1].reshape(count, count)
c = heat_data.T[2].reshape(count, count) / factor
c = c[:-1, :-1]

c[c==1] = 4.

l_a=a.min()
r_a=a.max()
l_b=b.min()
r_b=b.max()
l_c,r_c  = np.abs(c).min(), np.abs(c).max()
l_c = 0
r_c = 4

plt.rcParams.update({'font.size': 50})

figure, axes = plt.subplots(figsize=(15, 11))

c = axes.pcolormesh(a, b, c, cmap='hot', vmin=l_c, vmax=r_c)
axes.set_title(title, size=85)
axes.axis([l_a, r_a, l_b, r_b])
figure.colorbar(c)

plt.tight_layout()
plt.savefig(f'/Users/ruismacdonald/Research/Dreamer/rewards/RHM_optimal_phase2.pdf')
plt.show()
```
