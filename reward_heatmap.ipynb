```python
for buffer, file, phase in [("FIFO", "fifo", 2), ("LoFoV1", "v1", 2), ("LoFoV2", "v2", 2)]:
    with open(f"/Users/ruismacdonald/Downloads/rewards/{file}_1_phase{phase}/heat_data.npy", "rb") as f:
        heat_data = np.load(f, allow_pickle=True)
        
    heat_data[heat_data[:,2] == 1][:,2] = 4

    title=f"{buffer} - Phase {phase}"
        
    count = int(np.sqrt(heat_data.T[0].shape[0]))
    factor = 2.
    a = heat_data.T[0].reshape(count, count)
    b = heat_data.T[1].reshape(count, count)
    c = heat_data.T[2].reshape(count, count) / factor
    c = c[:-1, :-1]

    c[c==1] = 4.

    l_a=a.min()
    r_a=a.max()
    l_b=b.min()
    r_b=b.max()
    l_c,r_c  = np.abs(c).min(), np.abs(c).max()
    l_c = 0
    r_c = 4

    plt.rcParams.update({'font.size': 50})

    figure, axes = plt.subplots(figsize=(15, 11))

    c = axes.pcolormesh(a, b, c, cmap='hot', vmin=l_c, vmax=r_c)
    axes.set_title(title, size=85)
    axes.set_xlim(-0.18, 0.18)
    axes.set_ylim(-0.18, 0.18)
    axes.set_xticks([-0.1, 0.0, 0.1])
    axes.set_yticks([-0.1, 0.0, 0.1])
    figure.colorbar(c)

    plt.tight_layout()
    plt.savefig(f'/Users/ruismacdonald/Downloads/RHM_{file}_phase{phase}.pdf')
    plt.show()
```
